```python
import os
import pandas as pd
```

```python
complete_df=pd.DataFrame()
```

```python
for file in os.listdir(os.getcwd()):
    if file.endswith('.csv'):
        complete_df=complete_df.append(pd.read_csv(file))
```

```python
complete_df.to_csv('Complete_File.csv',index=False)
```
